# Read Dataset

```python
import os
import kagglehub

# Download latest version to define 'path'
path = kagglehub.dataset_download("kalkulasi/financial-statement-data-idx-2020-2023")

# List files in the downloaded directory
print(os.listdir(path))

# Construct the full path to the CSV file
csv_file_path = os.path.join(path, 'combined_financial_data_idx.csv')

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

print("Path to dataset files:", path)
display(df.head())
```

# Data Cleaning

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
# untuk melihat value kolom
for column in df.columns:
    print(f"============= {column} =================")
    display(df[column].value_counts())
    print()
```

```python
# 1Interpolasi antar tahun (untuk kolom numerik)
df[['2020', '2021', '2022', '2023']] = df[['2020', '2021', '2022', '2023']].interpolate(
    method='linear', axis=1)

# 2Forward fill / backward fill untuk nilai di awal / akhir
df[['2020', '2021', '2022', '2023']] = df[['2020', '2021', '2022', '2023']].fillna(method='ffill', axis=1)
df[['2020', '2021', '2022', '2023']] = df[['2020', '2021', '2022', '2023']].fillna(method='bfill', axis=1)

# 3Drop baris yang tetap NaN di semua tahun
df = df.dropna(subset=['2020', '2021', '2022', '2023'], how='all')

df.head()
```

- Interpolasi Linier (interpolate(method='linear', axis=1)): Mengisi nilai NaN di antara tahun-tahun yang memiliki data. Misalnya, jika data tahun 2020 dan 2022 ada tetapi 2021 hilang, interpolasi linier akan mengisi 2021 dengan nilai di tengah antara 2020 dan 2022. Ini berasumsi ada tren linier antara titik data yang diketahui.

- Forward Fill (fillna(method='ffill', axis=1)): Mengisi nilai NaN dengan nilai terakhir yang valid di baris yang sama. Berguna jika Anda berasumsi nilai keuangan tetap sama dari tahun sebelumnya jika tidak dilaporkan.

- Backward Fill (fillna(method='bfill', axis=1)): Mengisi nilai NaN dengan nilai valid pertama setelahnya di baris yang sama. Berguna untuk asumsi nilai keuangan akan sama dengan nilai yang dilaporkan di tahun berikutnya.

- Drop Rows (dropna(subset=..., how='all')): Setelah mencoba mengisi nilai, kode ini menghapus baris di mana semua nilai untuk tahun 2020 hingga 2023 masih NaN.

- Count (Jumlah Data): Semua kolom tahun sekarang memiliki jumlah data yang sama (89243), yang mengkonfirmasi bahwa nilai-nilai NaN telah berhasil ditangani dan tidak ada baris yang dihapus secara massal kecuali yang memang kosong di semua tahun.
Mean (Rata-rata): Nilai rata-rata menunjukkan tren kenaikan dari tahun 2020 hingga 2023 (dari sekitar 1.13 triliun menjadi 1.44 triliun). Ini mengindikasikan bahwa, secara rata-rata di seluruh akun keuangan dan perusahaan, nilai-nilai cenderung meningkat selama periode ini.
- Std (Standar Deviasi): Standar deviasi sangat besar dan meningkat dari tahun ke tahun. Ini menunjukkan variasi yang sangat besar dalam nilai-nilai keuangan di seluruh data Anda. Ada perbedaan yang signifikan antara nilai terendah dan tertinggi, serta banyak nilai yang tersebar jauh dari rata-rata. Ini wajar dalam data keuangan yang mencakup perusahaan besar dan kecil dengan akun keuangan yang sangat bervariasi nilainya.
Min (Nilai Minimum): Ada nilai minimum negatif yang cukup besar, menunjukkan bahwa ada akun keuangan (seperti akumulasi penyusutan atau beberapa jenis liabilitas/ekuitas) yang memang bernilai negatif. Nilai minimum ini juga cenderung menurun (menjadi lebih negatif) dari tahun ke tahun.
- 25% (Kuartil Pertama): Seperti yang kita diskusikan, 25% data terendah di setiap tahun bernilai 0.00. Ini menegaskan bahwa seperempat dari data memiliki nilai nol.
- 50% (Median/Kuartil Kedua): Nilai median (atau kuartil kedua) jauh lebih rendah daripada rata-rata (sekitar 4.6 miliar hingga 6.5 miliar). Ini menunjukkan bahwa distribusi data sangat miring (skewed) ke kanan, dengan banyak nilai yang relatif kecil dan sejumlah kecil nilai yang sangat besar (outlier) yang menarik rata-rata ke atas. Median memberikan gambaran yang lebih baik tentang "nilai tipikal" dalam data dibandingkan rata-rata. Nilai median juga menunjukkan sedikit peningkatan dari tahun ke tahun.
- 75% (Kuartil Ketiga): Nilai kuartil ketiga (sekitar 191 miliar hingga 257 miliar) jauh di atas median tetapi masih jauh di bawah rata-rata maksimum. Ini menunjukkan bahwa sebagian besar data (dari 25% hingga 75%) berada dalam rentang nilai yang lebih terkonsentrasi dibandingkan keseluruhan rentang data.
- Max (Nilai Maksimum): Ada nilai maksimum yang sangat besar (mencapai triliunan), yang merupakan outlier positif yang signifikan dan berkontribusi besar terhadap nilai rata-rata yang tinggi dan standar deviasi yang besar.

### Data Transformation for Visualization (Long Format)

```python
# Melt the DataFrame to transform it from wide to long format
# This is ideal for visualizing time-series data in Streamlit
df_long = df.melt(id_vars=['symbol', 'account', 'type'],
                  var_name='Year',
                  value_name='Value')

# Convert 'Year' column to integer type for proper sorting and plotting
df_long['Year'] = df_long['Year'].astype(int)

print("DataFrame in long format:")
display(df_long.head())
```
